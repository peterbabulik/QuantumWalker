```python
import cirq
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import time

# --- Parameters ---
N_SITES_1D = 51
# DEPTH_1D will be set per experiment in __main__
INITIAL_POS_1D = N_SITES_1D // 2
INITIAL_COIN_1D = 0
INITIAL_CA_CENTER_ONE = True

# --- Default Two-Way Coupling Parameters (can be overridden per experiment) ---
# P_THRESHOLD_DEFAULT = 0.1 # Will be set in the loop
# FEEDBACK_TYPE_DEFAULT = "flip" # Will be set in the loop

# --- CA Rule Update Function ---
def update_ca_line(current_line, rule_number):
    width = len(current_line)
    new_line = np.zeros(width, dtype=int)
    for i in range(width):
        prev_left = current_line[(i - 1 + width) % width]
        prev_center = current_line[i]
        prev_right = current_line[(i + 1) % width]
        pattern_val = (prev_left << 2) | (prev_center << 1) | prev_right
        if rule_number == 30:
            if pattern_val in [1, 2, 3, 4]: new_line[i] = 1
            else: new_line[i] = 0
        elif rule_number == 90:
            if prev_left != prev_right: new_line[i] = 1
            else: new_line[i] = 0
        elif rule_number == 110:
            if pattern_val in [1, 2, 3, 5, 6]: new_line[i] = 1
            else: new_line[i] = 0
        else: raise ValueError(f"Rule {rule_number} not implemented.")
    return new_line

# --- Coin Matrices (1-Qubit Coin) ---
H_1Q = (1/np.sqrt(2)) * np.array([[1, 1], [1, -1]], dtype=np.complex128)
X_1Q = np.array([[0,1],[1,0]], dtype=np.complex128)
I_1Q = np.eye(2, dtype=np.complex128) # DEFINITION ADDED

# These are defaults, actual ops passed to simulation will be from coin_op_pairs
COIN_OP_CA_ZERO_DEFAULT = H_1Q
COIN_OP_CA_ONE_DEFAULT  = X_1Q @ H_1Q

# --- Helper Functions (1D QW) ---
def get_1d_index(coin_val, site_pos, n_sites_1d):
    if not (0 <= site_pos < n_sites_1d and 0 <= coin_val <= 1):
        raise IndexError(f"Invalid coin ({coin_val}) or pos ({site_pos}) for 1D index (N_sites={n_sites_1d})")
    return coin_val + 2 * site_pos

def get_1d_coin_pos_from_index(k, n_sites_1d):
    state_dim_1d = 2 * n_sites_1d
    if not (0 <= k < state_dim_1d):
        raise IndexError(f"Invalid k ({k}) for 1D coin/pos (StateDim={state_dim_1d})")
    coin_val = k % 2
    site_pos = k // 2
    return coin_val, site_pos

# --- Initial State Preparation (1D QW) ---
def prepare_initial_state_1d(n_sites_1d, initial_pos, initial_coin_val=0):
    state_dim_1d = 2 * n_sites_1d
    initial_state_vector = np.zeros(state_dim_1d, dtype=np.complex128)
    try:
        start_index = get_1d_index(initial_coin_val, initial_pos, n_sites_1d)
        initial_state_vector[start_index] = 1.0
        # print(f"Initialized 1D walker at site {initial_pos} in coin state |{initial_coin_val}> (index {start_index})") # Reduce noise
    except IndexError as e:
        print(f"Error during 1D initial state prep: {e}")
    return initial_state_vector

# --- Build QW Step Operator ---
def build_1d_qw_step_operator(n_sites_1d, current_ca_pattern, coin0_op, coin1_op):
    state_dim_1d = 2 * n_sites_1d
    C_full = np.zeros((state_dim_1d, state_dim_1d), dtype=np.complex128)
    for site_idx in range(n_sites_1d):
        chosen_coin_op = coin0_op if current_ca_pattern[site_idx] == 0 else coin1_op
        idx_s_c0 = get_1d_index(0, site_idx, n_sites_1d)
        idx_s_c1 = get_1d_index(1, site_idx, n_sites_1d)
        C_full[idx_s_c0, idx_s_c0] = chosen_coin_op[0,0]
        C_full[idx_s_c1, idx_s_c0] = chosen_coin_op[1,0]
        C_full[idx_s_c0, idx_s_c1] = chosen_coin_op[0,1]
        C_full[idx_s_c1, idx_s_c1] = chosen_coin_op[1,1]
    S_full = np.zeros((state_dim_1d, state_dim_1d), dtype=np.complex128)
    for k_in in range(state_dim_1d):
        coin_val, site_pos = get_1d_coin_pos_from_index(k_in, n_sites_1d)
        new_pos = site_pos
        if coin_val == 0:
            new_pos = (site_pos - 1 + n_sites_1d) % n_sites_1d
        else:
            new_pos = (site_pos + 1) % n_sites_1d
        k_out = get_1d_index(coin_val, new_pos, n_sites_1d)
        S_full[k_out, k_in] = 1.0
    U_step = S_full @ C_full
    return U_step

# --- Observables (1D QW) ---
def calculate_p1_position_1d(state_vector, n_sites_1d):
    prob_dist_1d = np.zeros(n_sites_1d)
    for site_idx in range(n_sites_1d):
        idx_c0 = get_1d_index(0, site_idx, n_sites_1d)
        idx_c1 = get_1d_index(1, site_idx, n_sites_1d)
        prob_dist_1d[site_idx] = np.abs(state_vector[idx_c0])**2 + np.abs(state_vector[idx_c1])**2
    return prob_dist_1d

def calculate_coin_position_entanglement_1d(state_vector, n_sites_1d):
    coin_dim = 2
    state_dim_1d = coin_dim * n_sites_1d
    if len(state_vector) != state_dim_1d: return np.nan
    norm = np.linalg.norm(state_vector)
    if np.abs(norm - 1.0) > 1e-6:
        if norm < 1e-9: return 0.0
        state_vector = state_vector / norm
        if np.isnan(state_vector).any(): return np.nan
    try:
        rho_full_matrix = np.outer(state_vector, np.conjugate(state_vector))
        rho_coin = np.zeros((coin_dim, coin_dim), dtype=np.complex128)
        for site_idx in range(n_sites_1d):
            for c_row in range(coin_dim):
                 for c_col in range(coin_dim):
                      k_row = get_1d_index(c_row, site_idx, n_sites_1d)
                      k_col = get_1d_index(c_col, site_idx, n_sites_1d)
                      rho_coin[c_row, c_col] += rho_full_matrix[k_row, k_col]
        trace_rho_coin = np.trace(rho_coin)
        if np.abs(trace_rho_coin - 1.0) > 1e-6:
            if abs(trace_rho_coin) > 1e-9: rho_coin /= trace_rho_coin
        eigenvalues = np.linalg.eigvalsh(rho_coin)
        entropy = 0.0; epsilon = 1e-12
        for eig_val in eigenvalues:
            if eig_val > epsilon:
                try: entropy -= eig_val * np.log2(eig_val)
                except ValueError: pass
        return max(0.0, np.real(entropy))
    except Exception as e:
        # print(f"Error calculating 1D coin-position entanglement: {e}") # Reduce noise
        return np.nan

# --- Simulation Loop (1D QW with Two-Way Coupled Dynamic CA Coin) ---
def run_1d_qw_two_way_coupling(
    n_sites_1d, depth, initial_pos, initial_coin_val,
    initial_ca_center_one, ca_rule_number, coin0_op, coin1_op,
    p_threshold, feedback_type
):
    current_qw_state = prepare_initial_state_1d(n_sites_1d, initial_pos, initial_coin_val)
    current_ca_line = np.zeros(n_sites_1d, dtype=int)
    if initial_ca_center_one and n_sites_1d > 0:
        if n_sites_1d % 2 == 1: current_ca_line[n_sites_1d // 2] = 1
        else: current_ca_line[n_sites_1d // 2 -1] = 1

    prob_history_1d = [calculate_p1_position_1d(current_qw_state, n_sites_1d)]
    ent_history_1d = [calculate_coin_position_entanglement_1d(current_qw_state, n_sites_1d)]
    ca_history = [current_ca_line.copy()]

    # print(f"\nStarting 1D QW (Two-Way Coupled Rule {ca_rule_number} Coin, P_thresh={p_threshold}, Feedback='{feedback_type}') for {depth} steps...") # Reduce noise
    start_time = time.time()
    for step in range(depth):
        U_step = build_1d_qw_step_operator(n_sites_1d, current_ca_line, coin0_op, coin1_op)
        current_qw_state = U_step @ current_qw_state
        norm_qw = np.linalg.norm(current_qw_state)
        if abs(norm_qw) > 1e-9 : current_qw_state /= norm_qw
        else: print(f"Warning: QW State norm zero at step {step+1}"); break

        qw_probs_at_sites = calculate_p1_position_1d(current_qw_state, n_sites_1d)
        ca_line_for_next_update = current_ca_line.copy()

        for site_idx in range(n_sites_1d):
            if qw_probs_at_sites[site_idx] > p_threshold:
                if feedback_type == "flip":
                    ca_line_for_next_update[site_idx] = 1 - ca_line_for_next_update[site_idx]
                elif feedback_type == "set_to_one":
                    ca_line_for_next_update[site_idx] = 1
                elif feedback_type == "set_to_zero":
                    ca_line_for_next_update[site_idx] = 0

        current_ca_line = update_ca_line(ca_line_for_next_update, ca_rule_number)

        prob_history_1d.append(qw_probs_at_sites)
        ent_history_1d.append(calculate_coin_position_entanglement_1d(current_qw_state, n_sites_1d))
        ca_history.append(current_ca_line.copy())

        if (step + 1) % (depth // 10 if depth >=50 else 5) == 0 :
            # print(f"  Completed step {step + 1}/{depth}") # Reduce noise
            pass
    end_time = time.time()
    # print(f"Evolution complete. Time: {end_time - start_time:.2f} seconds.") # Reduce noise

    return {
        "prob_history": np.array(prob_history_1d),
        "ent_history": np.array(ent_history_1d),
        "ca_history": np.array(ca_history),
        "params": {
            "n_sites": n_sites_1d, "depth": depth,
            "initial_pos": initial_pos, "initial_coin": initial_coin_val,
            "initial_ca_center_one": initial_ca_center_one,
            "ca_rule_number": ca_rule_number,
            "p_threshold": p_threshold,
            "feedback_type": feedback_type
        }
    }

# --- Plotting and Summary (1D QW with Dynamic CA and Two-Way Coupling) ---
def plot_1d_qw_results_two_way(results):
    params = results["params"]; prob_hist = results["prob_history"]
    ent_hist = results["ent_history"]; ca_hist = results["ca_history"]
    depth = params['depth']; ca_rule_number = params['ca_rule_number']
    p_thresh = params['p_threshold']; feedback = params['feedback_type']
    c0_name = params.get("coin0_name", "C0_def")
    c1_name = params.get("coin1_name", "C1_def")

    fig, axs = plt.subplots(4, 1, figsize=(12, 14), gridspec_kw={'height_ratios': [1, 3, 3, 2]})
    title_str = (f"1D QW Two-Way Coupled with Rule {ca_rule_number} CA\n"
                 f"(N={params['n_sites']}, D={depth}, Pth={p_thresh}, Fb='{feedback}', C0='{c0_name}', C1='{c1_name}')")
    fig.suptitle(title_str, fontsize=12)

    axs[0].imshow(ca_hist, cmap='binary', aspect='auto', interpolation='nearest')
    axs[0].set_title(f"Dynamic Rule {ca_rule_number} Pattern (Influenced by QW)")
    axs[0].set_xlabel("Site Index"); axs[0].set_yticks([])

    axs[1].plot(np.arange(params['n_sites']), prob_hist[-1, :], marker='o', linestyle='-', color='blue')
    axs[1].set_title("Final QW Probability Distribution P(x)")
    axs[1].set_xlabel("Site Index (x)"); axs[1].set_ylabel("Probability P(x)")
    axs[1].grid(True); axs[1].set_ylim(bottom=-0.01)

    prob_spacetime = prob_hist.T
    pos_probs = prob_spacetime[prob_spacetime > 1e-9]
    min_val = pos_probs.min() if len(pos_probs) > 0 else 1e-9
    max_val = prob_spacetime.max()
    norm_val = colors.LogNorm(vmin=min_val, vmax=max(max_val, min_val + 1e-6))
    im = axs[2].imshow(prob_spacetime, aspect='auto', origin='lower', cmap='magma', norm=norm_val,
                       extent=[0, depth, 0, params['n_sites']-1])
    plt.colorbar(im, ax=axs[2], label="P(x,t) (log scale)")
    axs[2].set_title("QW Probability Spacetime P(x,t)")
    axs[2].set_xlabel("QW Time Step (t)"); axs[2].set_ylabel("Site Index (x)")

    time_steps_qw = np.arange(depth + 1)
    axs[3].plot(time_steps_qw, ent_hist, marker='.', linestyle='-', color='green')
    axs[3].set_title("Coin-Position Entanglement Entropy S(t)")
    axs[3].set_xlabel("QW Time Step (t)"); axs[3].set_ylabel("Entropy S")
    axs[3].grid(True); axs[3].set_ylim(bottom=-0.01, top=1.05)

    plt.tight_layout(rect=[0, 0, 1, 0.92]);
    plt.show()

def generate_1d_qw_text_output_two_way(results):
    params = results["params"]; prob_hist = results["prob_history"]; ent_hist = results["ent_history"]
    ca_rule_number = params['ca_rule_number']
    p_thresh = params['p_threshold']; feedback = params['feedback_type']
    c0_name = params.get("coin0_name", "C0_def")
    c1_name = params.get("coin1_name", "C1_def")

    output_lines = [f"\n--- 1D QW (Two-Way Coupled Rule {ca_rule_number}, Pth={p_thresh}, Fb='{feedback}', C0='{c0_name}', C1='{c1_name}') Summary ---"]
    output_lines.append(f"N_sites={params['n_sites']}, Depth={params['depth']}, StartPos={params['initial_pos']}, StartCoin={params['initial_coin']}")

    output_lines.append("\nTop 5 Final Probabilities (Site: Prob):")
    final_probs = prob_hist[-1, :]
    sorted_indices = np.argsort(final_probs)[::-1]
    for i in range(min(5, params['n_sites'])):
        idx = sorted_indices[i]
        output_lines.append(f"  Site {idx}: {final_probs[idx]:.5f}")

    output_lines.append("\nCoin-Position Entanglement vs. Time:")
    num_steps_total = len(ent_hist)
    indices_to_show = [0]
    if num_steps_total > 2: indices_to_show.append(num_steps_total // 2)
    if num_steps_total > 1: indices_to_show.append(num_steps_total - 1)
    indices_to_show = sorted(list(set(indices_to_show)))
    for idx in indices_to_show:
        s_val_str = f"{ent_hist[idx]:.6f}" if np.isfinite(ent_hist[idx]) else "nan"
        output_lines.append(f"  Step {idx}: S = {s_val_str}")
    output_lines.append("--- End of Summary ---")
    return "\n".join(output_lines)


# --- Main Execution (1D QW - Particle Search) ---
if __name__ == "__main__":

    all_results_text = []
    experiment_counter = 1

    # --- Define Search Space ---
    ca_rules_to_search = [110]

    coin_op_pairs = [
        (H_1Q, X_1Q @ H_1Q),
        (H_1Q, H_1Q),
        (H_1Q, I_1Q),
        (H_1Q, X_1Q),
        (I_1Q, H_1Q),
        (X_1Q, H_1Q),
        (I_1Q, I_1Q)
    ]

    p_thresholds_to_search = [0.01, 0.05, 0.1, 0.3] # Added 0.01 and expanded range
    feedback_types_to_search = ["flip", "set_to_one", "set_to_zero"]

    search_depth = 200 # Default depth for these search runs, can override per experiment

    # --- Experiment A: Longer Evolution with Rule 110 (as before) ---
    print(f"\n===== EXPERIMENT {experiment_counter}: Rule 110 - Long Evolution (Two-Way) =====")
    results_A = run_1d_qw_two_way_coupling(
        n_sites_1d=N_SITES_1D, depth=500,
        initial_pos=INITIAL_POS_1D, initial_coin_val=INITIAL_COIN_1D,
        initial_ca_center_one=INITIAL_CA_CENTER_ONE, ca_rule_number=110,
        coin0_op=H_1Q, coin1_op=X_1Q @ H_1Q, # Using a specific coin pair for this long run
        p_threshold=0.05, feedback_type="flip"
    )
    results_A["params"]["coin0_name"] = "H"; results_A["params"]["coin1_name"] = "XH" # Add coin names
    text_summary_A = generate_1d_qw_text_output_two_way(results_A)
    all_results_text.append(f"--- Experiment {experiment_counter} Results ---\n{text_summary_A}")
    print(text_summary_A); plot_1d_qw_results_two_way(results_A)
    print(f"===== Finished Experiment {experiment_counter} =====")
    experiment_counter += 1

    # --- Experiment B & C Combined: Sweep P_THRESH, FEEDBACK_TYPE, COIN_PAIRS for Rule 110 ---
    # This will be a large number of runs, be mindful of execution time.
    # Consider selecting a smaller subset for initial exploration.
    # For now, let's just do one P_thresh and one feedback type with all coin pairs.

    selected_p_thresh = 0.05
    selected_feedback_type = "flip"
    current_search_depth = 200 # Shorter for this sweep

    print(f"\n===== STARTING RULE 110 SWEEP: P_thresh={selected_p_thresh}, Feedback='{selected_feedback_type}', Depth={current_search_depth} =====")
    for idx_pair, (c0_op, c1_op) in enumerate(coin_op_pairs):
        def get_op_name(op_matrix):
            if np.allclose(op_matrix, H_1Q): return "H"
            if np.allclose(op_matrix, X_1Q): return "X"
            if np.allclose(op_matrix, I_1Q): return "I"
            if np.allclose(op_matrix, X_1Q @ H_1Q): return "XH"
            return "Custom"
        c0_name = get_op_name(c0_op); c1_name = get_op_name(c1_op)

        experiment_title = f"Rule 110, Pth={selected_p_thresh}, Fb='{selected_feedback_type}', C0='{c0_name}', C1='{c1_name}'"
        print(f"\n===== EXPERIMENT {experiment_counter}: {experiment_title} =====")

        results_sweep = run_1d_qw_two_way_coupling(
            n_sites_1d=N_SITES_1D, depth=current_search_depth,
            initial_pos=INITIAL_POS_1D, initial_coin_val=INITIAL_COIN_1D,
            initial_ca_center_one=INITIAL_CA_CENTER_ONE, ca_rule_number=110,
            coin0_op=c0_op, coin1_op=c1_op,
            p_threshold=selected_p_thresh, feedback_type=selected_feedback_type
        )
        results_sweep["params"]["coin0_name"] = c0_name
        results_sweep["params"]["coin1_name"] = c1_name
        text_summary_sweep = generate_1d_qw_text_output_two_way(results_sweep)
        all_results_text.append(f"--- Experiment {experiment_counter} ({experiment_title}) ---\n{text_summary_sweep}")
        print(text_summary_sweep); plot_1d_qw_results_two_way(results_sweep)
        print(f"===== Finished Experiment {experiment_counter} =====")
        experiment_counter += 1
    print(f"===== FINISHED RULE 110 SWEEP =====")


    # --- Comparison Runs for Rule 90 and Rule 30 ---
    comparison_rules_depth = 100
    comparison_p_thresh = 0.1
    comparison_feedback = "flip"
    # Using the "standard" coin ops (H, XH) for these comparison runs
    comp_c0_op = H_1Q; comp_c1_op = X_1Q @ H_1Q
    comp_c0_name = get_op_name(comp_c0_op); comp_c1_name = get_op_name(comp_c1_op)


    for rule_num_comp in [90, 30]:
        experiment_title_comp = f"Rule {rule_num_comp}, Pth={comparison_p_thresh}, Fb='{comparison_feedback}', C0='{comp_c0_name}', C1='{comp_c1_name}'"
        print(f"\n===== EXPERIMENT {experiment_counter}: {experiment_title_comp} (Comparison) =====")
        results_Comp = run_1d_qw_two_way_coupling(
            n_sites_1d=N_SITES_1D, depth=comparison_rules_depth,
            initial_pos=INITIAL_POS_1D, initial_coin_val=INITIAL_COIN_1D,
            initial_ca_center_one=INITIAL_CA_CENTER_ONE, ca_rule_number=rule_num_comp,
            coin0_op=comp_c0_op, coin1_op=comp_c1_op,
            p_threshold=comparison_p_thresh, feedback_type=comparison_feedback
        )
        results_Comp["params"]["coin0_name"] = comp_c0_name
        results_Comp["params"]["coin1_name"] = comp_c1_name
        text_summary_Comp = generate_1d_qw_text_output_two_way(results_Comp)
        all_results_text.append(f"--- Experiment {experiment_counter} ({experiment_title_comp}) ---\n{text_summary_Comp}")
        print(text_summary_Comp)
        plot_1d_qw_results_two_way(results_Comp)
        print(f"===== Finished Experiment {experiment_counter} (Rule {rule_num_comp}) =====")
        experiment_counter += 1


    print("\n\n<<<<< ALL SIMULATION SUMMARIES (Particle Search & Comparisons - Two-Way Coupled) >>>>>")
    for summary in all_results_text:
        print(summary)
    print("<<<<< END OF ALL SIMULATION SUMMARIES >>>>>")
```

```python
import cirq
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import time

# --- Parameters ---
N_SITES_1D = 51
# DEPTH_1D will be set in __main__ for the specific long run
INITIAL_POS_1D = N_SITES_1D // 2
INITIAL_COIN_1D = 0
INITIAL_CA_CENTER_ONE = True

# --- Default Two-Way Coupling Parameters ---
# P_THRESHOLD and FEEDBACK_TYPE will be set in __main__

# --- CA Rule Update Function ---
def update_ca_line(current_line, rule_number):
    width = len(current_line)
    new_line = np.zeros(width, dtype=int)
    for i in range(width):
        prev_left = current_line[(i - 1 + width) % width]
        prev_center = current_line[i]
        prev_right = current_line[(i + 1) % width]
        pattern_val = (prev_left << 2) | (prev_center << 1) | prev_right
        if rule_number == 30:
            if pattern_val in [1, 2, 3, 4]: new_line[i] = 1
            else: new_line[i] = 0
        elif rule_number == 90:
            if prev_left != prev_right: new_line[i] = 1
            else: new_line[i] = 0
        elif rule_number == 110:
            if pattern_val in [1, 2, 3, 5, 6]: new_line[i] = 1
            else: new_line[i] = 0
        else: raise ValueError(f"Rule {rule_number} not implemented.")
    return new_line

# --- Coin Matrices (1-Qubit Coin) ---
H_1Q = (1/np.sqrt(2)) * np.array([[1, 1], [1, -1]], dtype=np.complex128)
X_1Q = np.array([[0,1],[1,0]], dtype=np.complex128)
I_1Q = np.eye(2, dtype=np.complex128)

# Define the specific pair we'll use for the long run
COIN_OP_R110_ZERO_LONG_RUN = H_1Q
COIN_OP_R110_ONE_LONG_RUN  = X_1Q @ H_1Q

# --- Helper Functions (1D QW) ---
def get_1d_index(coin_val, site_pos, n_sites_1d):
    if not (0 <= site_pos < n_sites_1d and 0 <= coin_val <= 1):
        raise IndexError(f"Invalid coin ({coin_val}) or pos ({site_pos}) for 1D index (N_sites={n_sites_1d})")
    return coin_val + 2 * site_pos

def get_1d_coin_pos_from_index(k, n_sites_1d):
    state_dim_1d = 2 * n_sites_1d
    if not (0 <= k < state_dim_1d):
        raise IndexError(f"Invalid k ({k}) for 1D coin/pos (StateDim={state_dim_1d})")
    coin_val = k % 2
    site_pos = k // 2
    return coin_val, site_pos

# --- Initial State Preparation (1D QW) ---
def prepare_initial_state_1d(n_sites_1d, initial_pos, initial_coin_val=0):
    state_dim_1d = 2 * n_sites_1d
    initial_state_vector = np.zeros(state_dim_1d, dtype=np.complex128)
    try:
        start_index = get_1d_index(initial_coin_val, initial_pos, n_sites_1d)
        initial_state_vector[start_index] = 1.0
        print(f"Initialized 1D walker at site {initial_pos} in coin state |{initial_coin_val}> (index {start_index})")
    except IndexError as e:
        print(f"Error during 1D initial state prep: {e}")
    return initial_state_vector

# --- Build QW Step Operator ---
def build_1d_qw_step_operator(n_sites_1d, current_ca_pattern, coin0_op, coin1_op):
    state_dim_1d = 2 * n_sites_1d
    C_full = np.zeros((state_dim_1d, state_dim_1d), dtype=np.complex128)
    for site_idx in range(n_sites_1d):
        chosen_coin_op = coin0_op if current_ca_pattern[site_idx] == 0 else coin1_op
        idx_s_c0 = get_1d_index(0, site_idx, n_sites_1d)
        idx_s_c1 = get_1d_index(1, site_idx, n_sites_1d)
        C_full[idx_s_c0, idx_s_c0] = chosen_coin_op[0,0]
        C_full[idx_s_c1, idx_s_c0] = chosen_coin_op[1,0]
        C_full[idx_s_c0, idx_s_c1] = chosen_coin_op[0,1]
        C_full[idx_s_c1, idx_s_c1] = chosen_coin_op[1,1]
    S_full = np.zeros((state_dim_1d, state_dim_1d), dtype=np.complex128)
    for k_in in range(state_dim_1d):
        coin_val, site_pos = get_1d_coin_pos_from_index(k_in, n_sites_1d)
        new_pos = site_pos
        if coin_val == 0:
            new_pos = (site_pos - 1 + n_sites_1d) % n_sites_1d
        else:
            new_pos = (site_pos + 1) % n_sites_1d
        k_out = get_1d_index(coin_val, new_pos, n_sites_1d)
        S_full[k_out, k_in] = 1.0
    U_step = S_full @ C_full
    return U_step

# --- Observables (1D QW) ---
def calculate_p1_position_1d(state_vector, n_sites_1d):
    prob_dist_1d = np.zeros(n_sites_1d)
    for site_idx in range(n_sites_1d):
        idx_c0 = get_1d_index(0, site_idx, n_sites_1d)
        idx_c1 = get_1d_index(1, site_idx, n_sites_1d)
        prob_dist_1d[site_idx] = np.abs(state_vector[idx_c0])**2 + np.abs(state_vector[idx_c1])**2
    return prob_dist_1d

def calculate_coin_position_entanglement_1d(state_vector, n_sites_1d):
    coin_dim = 2
    state_dim_1d = coin_dim * n_sites_1d
    if len(state_vector) != state_dim_1d: return np.nan
    norm = np.linalg.norm(state_vector)
    if np.abs(norm - 1.0) > 1e-6:
        if norm < 1e-9: return 0.0
        state_vector = state_vector / norm
        if np.isnan(state_vector).any(): return np.nan
    try:
        rho_full_matrix = np.outer(state_vector, np.conjugate(state_vector))
        rho_coin = np.zeros((coin_dim, coin_dim), dtype=np.complex128)
        for site_idx in range(n_sites_1d):
            for c_row in range(coin_dim):
                 for c_col in range(coin_dim):
                      k_row = get_1d_index(c_row, site_idx, n_sites_1d)
                      k_col = get_1d_index(c_col, site_idx, n_sites_1d)
                      rho_coin[c_row, c_col] += rho_full_matrix[k_row, k_col]
        trace_rho_coin = np.trace(rho_coin)
        if np.abs(trace_rho_coin - 1.0) > 1e-6:
            if abs(trace_rho_coin) > 1e-9: rho_coin /= trace_rho_coin
        eigenvalues = np.linalg.eigvalsh(rho_coin)
        entropy = 0.0; epsilon = 1e-12
        for eig_val in eigenvalues:
            if eig_val > epsilon:
                try: entropy -= eig_val * np.log2(eig_val)
                except ValueError: pass
        return max(0.0, np.real(entropy))
    except Exception as e:
        # print(f"Error calculating 1D coin-position entanglement: {e}") # Reduce noise
        return np.nan

# --- Simulation Loop (1D QW with Two-Way Coupled Dynamic CA Coin) ---
def run_1d_qw_two_way_coupling(
    n_sites_1d, depth, initial_pos, initial_coin_val,
    initial_ca_center_one, ca_rule_number, coin0_op, coin1_op,
    p_threshold, feedback_type
):
    current_qw_state = prepare_initial_state_1d(n_sites_1d, initial_pos, initial_coin_val)
    current_ca_line = np.zeros(n_sites_1d, dtype=int)
    if initial_ca_center_one and n_sites_1d > 0:
        if n_sites_1d % 2 == 1: current_ca_line[n_sites_1d // 2] = 1
        else: current_ca_line[n_sites_1d // 2 -1] = 1

    prob_history_1d = [calculate_p1_position_1d(current_qw_state, n_sites_1d)]
    ent_history_1d = [calculate_coin_position_entanglement_1d(current_qw_state, n_sites_1d)]
    ca_history = [current_ca_line.copy()]

    print_feedback_info = True # Flag to control feedback summary print

    # print(f"\nStarting 1D QW (Two-Way Coupled Rule {ca_rule_number} Coin, P_thresh={p_threshold}, Feedback='{feedback_type}') for {depth} steps...")
    start_time = time.time()
    for step in range(depth):
        U_step = build_1d_qw_step_operator(n_sites_1d, current_ca_line, coin0_op, coin1_op)
        current_qw_state = U_step @ current_qw_state
        norm_qw = np.linalg.norm(current_qw_state)
        if abs(norm_qw) > 1e-9 : current_qw_state /= norm_qw
        else: print(f"Warning: QW State norm zero at step {step+1}"); break

        qw_probs_at_sites = calculate_p1_position_1d(current_qw_state, n_sites_1d)
        ca_line_for_next_update = current_ca_line.copy()

        num_influenced_cells_this_step = 0
        for site_idx in range(n_sites_1d):
            if qw_probs_at_sites[site_idx] > p_threshold:
                num_influenced_cells_this_step +=1
                if feedback_type == "flip":
                    ca_line_for_next_update[site_idx] = 1 - ca_line_for_next_update[site_idx]
                elif feedback_type == "set_to_one":
                    ca_line_for_next_update[site_idx] = 1
                elif feedback_type == "set_to_zero":
                    ca_line_for_next_update[site_idx] = 0

        current_ca_line = update_ca_line(ca_line_for_next_update, ca_rule_number)

        prob_history_1d.append(qw_probs_at_sites)
        ent_history_1d.append(calculate_coin_position_entanglement_1d(current_qw_state, n_sites_1d))
        ca_history.append(current_ca_line.copy())

        if (step + 1) % (depth // 20 if depth >=100 else 10) == 0 : # Adjusted print frequency
            print(f"  Step {step + 1}/{depth} completed. CA cells influenced in this step: {num_influenced_cells_this_step}")

    end_time = time.time()
    print(f"Evolution complete. Time: {end_time - start_time:.2f} seconds.")

    return {
        "prob_history": np.array(prob_history_1d),
        "ent_history": np.array(ent_history_1d),
        "ca_history": np.array(ca_history),
        "params": {
            "n_sites": n_sites_1d, "depth": depth,
            "initial_pos": initial_pos, "initial_coin": initial_coin_val,
            "initial_ca_center_one": initial_ca_center_one,
            "ca_rule_number": ca_rule_number,
            "p_threshold": p_threshold,
            "feedback_type": feedback_type
        }
    }

# --- Plotting and Summary (1D QW with Dynamic CA and Two-Way Coupling) ---
def plot_1d_qw_results_two_way(results):
    params = results["params"]; prob_hist = results["prob_history"]
    ent_hist = results["ent_history"]; ca_hist = results["ca_history"]
    depth = params['depth']; ca_rule_number = params['ca_rule_number']
    p_thresh = params['p_threshold']; feedback = params['feedback_type']
    c0_name = params.get("coin0_name", "C0_def")
    c1_name = params.get("coin1_name", "C1_def")

    fig, axs = plt.subplots(4, 1, figsize=(12, 14), gridspec_kw={'height_ratios': [1, 3, 3, 2]})
    title_str = (f"1D QW Two-Way Coupled with Rule {ca_rule_number} CA\n"
                 f"(N={params['n_sites']}, D={depth}, Pth={p_thresh}, Fb='{feedback}', C0='{c0_name}', C1='{c1_name}')")
    fig.suptitle(title_str, fontsize=12)

    axs[0].imshow(ca_hist, cmap='binary', aspect='auto', interpolation='nearest')
    axs[0].set_title(f"Dynamic Rule {ca_rule_number} Pattern (Influenced by QW)")
    axs[0].set_xlabel("Site Index"); axs[0].set_yticks([])

    axs[1].plot(np.arange(params['n_sites']), prob_hist[-1, :], marker='o', linestyle='-', color='blue')
    axs[1].set_title("Final QW Probability Distribution P(x)")
    axs[1].set_xlabel("Site Index (x)"); axs[1].set_ylabel("Probability P(x)")
    axs[1].grid(True); axs[1].set_ylim(bottom=-0.01)

    prob_spacetime = prob_hist.T
    pos_probs = prob_spacetime[prob_spacetime > 1e-9]
    min_val = pos_probs.min() if len(pos_probs) > 0 else 1e-9
    max_val = prob_spacetime.max()
    norm_val = colors.LogNorm(vmin=min_val, vmax=max(max_val, min_val + 1e-6))
    im = axs[2].imshow(prob_spacetime, aspect='auto', origin='lower', cmap='magma', norm=norm_val,
                       extent=[0, depth, 0, params['n_sites']-1])
    plt.colorbar(im, ax=axs[2], label="P(x,t) (log scale)")
    axs[2].set_title("QW Probability Spacetime P(x,t)")
    axs[2].set_xlabel("QW Time Step (t)"); axs[2].set_ylabel("Site Index (x)")

    time_steps_qw = np.arange(depth + 1)
    axs[3].plot(time_steps_qw, ent_hist, marker='.', linestyle='-', color='green')
    axs[3].set_title("Coin-Position Entanglement Entropy S(t)")
    axs[3].set_xlabel("QW Time Step (t)"); axs[3].set_ylabel("Entropy S")
    axs[3].grid(True); axs[3].set_ylim(bottom=-0.01, top=1.05)

    plt.tight_layout(rect=[0, 0, 1, 0.92]);
    plt.show()

def generate_1d_qw_text_output_two_way(results):
    params = results["params"]; prob_hist = results["prob_history"]; ent_hist = results["ent_history"]
    ca_rule_number = params['ca_rule_number']
    p_thresh = params['p_threshold']; feedback = params['feedback_type']
    c0_name = params.get("coin0_name", "C0_def")
    c1_name = params.get("coin1_name", "C1_def")

    output_lines = [f"\n--- 1D QW (Two-Way Coupled Rule {ca_rule_number}, Pth={p_thresh}, Fb='{feedback}', C0='{c0_name}', C1='{c1_name}') Summary ---"]
    output_lines.append(f"N_sites={params['n_sites']}, Depth={params['depth']}, StartPos={params['initial_pos']}, StartCoin={params['initial_coin']}")

    output_lines.append("\nTop 5 Final Probabilities (Site: Prob):")
    final_probs = prob_hist[-1, :]
    sorted_indices = np.argsort(final_probs)[::-1]
    for i in range(min(5, params['n_sites'])):
        idx = sorted_indices[i]
        output_lines.append(f"  Site {idx}: {final_probs[idx]:.5f}")

    output_lines.append("\nCoin-Position Entanglement vs. Time:")
    num_steps_total = len(ent_hist)
    indices_to_show = [0]
    if num_steps_total > 2: indices_to_show.append(num_steps_total // 2)
    if num_steps_total > 1: indices_to_show.append(num_steps_total - 1)
    indices_to_show = sorted(list(set(indices_to_show)))
    for idx in indices_to_show:
        s_val_str = f"{ent_hist[idx]:.6f}" if np.isfinite(ent_hist[idx]) else "nan"
        output_lines.append(f"  Step {idx}: S = {s_val_str}")
    output_lines.append("--- End of Summary ---")
    return "\n".join(output_lines)


# --- Main Execution (1D QW - Long Run Rule 110) ---
if __name__ == "__main__":

    all_results_text = []

    # --- Experiment: Long Evolution with Rule 110 ---
    long_run_depth = 5000
    long_run_p_thresh = 0.05
    long_run_feedback = "flip"
    long_run_rule = 110
    long_run_c0 = H_1Q
    long_run_c1 = X_1Q @ H_1Q

    def get_op_name(op_matrix): # Helper to get coin names for plot/summary
        if np.allclose(op_matrix, H_1Q): return "H"
        if np.allclose(op_matrix, X_1Q): return "X"
        if np.allclose(op_matrix, I_1Q): return "I"
        if np.allclose(op_matrix, X_1Q @ H_1Q): return "XH"
        return "Custom"

    c0_name_long = get_op_name(long_run_c0)
    c1_name_long = get_op_name(long_run_c1)

    experiment_title = f"Rule {long_run_rule}, D={long_run_depth}, Pth={long_run_p_thresh}, Fb='{long_run_feedback}', C0='{c0_name_long}', C1='{c1_name_long}'"
    print(f"\n===== STARTING LONG RUN EXPERIMENT: {experiment_title} =====")

    results_long_run = run_1d_qw_two_way_coupling(
        n_sites_1d=N_SITES_1D,
        depth=long_run_depth,
        initial_pos=INITIAL_POS_1D,
        initial_coin_val=INITIAL_COIN_1D,
        initial_ca_center_one=INITIAL_CA_CENTER_ONE,
        ca_rule_number=long_run_rule,
        coin0_op=long_run_c0,
        coin1_op=long_run_c1,
        p_threshold=long_run_p_thresh,
        feedback_type=long_run_feedback
    )
    results_long_run["params"]["coin0_name"] = c0_name_long # Add coin names for plotting
    results_long_run["params"]["coin1_name"] = c1_name_long

    text_summary_long = generate_1d_qw_text_output_two_way(results_long_run)
    all_results_text.append(f"--- Long Run Experiment Results ---\n{text_summary_long}")
    print(text_summary_long)
    plot_1d_qw_results_two_way(results_long_run)
    print(f"===== Finished Long Run Experiment =====")


    print("\n\n<<<<< ALL SIMULATION SUMMARIES >>>>>")
    for summary in all_results_text:
        print(summary)
    print("<<<<< END OF ALL SIMULATION SUMMARIES >>>>>")
```

```python
import cirq
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import time

# --- Parameters ---
N_SITES_1D = 51
# DEPTH_1D will be set in __main__ for the specific long run
INITIAL_POS_1D = N_SITES_1D // 2
INITIAL_COIN_1D = 0
INITIAL_CA_CENTER_ONE = True

# --- Default Two-Way Coupling Parameters ---
# P_THRESHOLD and FEEDBACK_TYPE will be set in __main__

# --- CA Rule Update Function ---
def update_ca_line(current_line, rule_number):
    width = len(current_line)
    new_line = np.zeros(width, dtype=int)
    for i in range(width):
        prev_left = current_line[(i - 1 + width) % width]
        prev_center = current_line[i]
        prev_right = current_line[(i + 1) % width]
        pattern_val = (prev_left << 2) | (prev_center << 1) | prev_right
        if rule_number == 30:
            if pattern_val in [1, 2, 3, 4]: new_line[i] = 1
            else: new_line[i] = 0
        elif rule_number == 90:
            if prev_left != prev_right: new_line[i] = 1
            else: new_line[i] = 0
        elif rule_number == 110:
            if pattern_val in [1, 2, 3, 5, 6]: new_line[i] = 1
            else: new_line[i] = 0
        else: raise ValueError(f"Rule {rule_number} not implemented.")
    return new_line

# --- Coin Matrices (1-Qubit Coin) ---
H_1Q = (1/np.sqrt(2)) * np.array([[1, 1], [1, -1]], dtype=np.complex128)
X_1Q = np.array([[0,1],[1,0]], dtype=np.complex128)
I_1Q = np.eye(2, dtype=np.complex128)

# --- Helper Functions (1D QW) ---
def get_1d_index(coin_val, site_pos, n_sites_1d):
    if not (0 <= site_pos < n_sites_1d and 0 <= coin_val <= 1):
        raise IndexError(f"Invalid coin ({coin_val}) or pos ({site_pos}) for 1D index (N_sites={n_sites_1d})")
    return coin_val + 2 * site_pos

def get_1d_coin_pos_from_index(k, n_sites_1d):
    state_dim_1d = 2 * n_sites_1d
    if not (0 <= k < state_dim_1d):
        raise IndexError(f"Invalid k ({k}) for 1D coin/pos (StateDim={state_dim_1d})")
    coin_val = k % 2
    site_pos = k // 2
    return coin_val, site_pos

# --- Initial State Preparation (1D QW) ---
def prepare_initial_state_1d(n_sites_1d, initial_pos, initial_coin_val=0):
    state_dim_1d = 2 * n_sites_1d
    initial_state_vector = np.zeros(state_dim_1d, dtype=np.complex128)
    try:
        start_index = get_1d_index(initial_coin_val, initial_pos, n_sites_1d)
        initial_state_vector[start_index] = 1.0
        # print(f"Initialized 1D walker at site {initial_pos} in coin state |{initial_coin_val}> (index {start_index})") # Reduce noise
    except IndexError as e:
        print(f"Error during 1D initial state prep: {e}")
    return initial_state_vector

# --- Build QW Step Operator ---
def build_1d_qw_step_operator(n_sites_1d, current_ca_pattern, coin0_op, coin1_op):
    state_dim_1d = 2 * n_sites_1d
    C_full = np.zeros((state_dim_1d, state_dim_1d), dtype=np.complex128)
    for site_idx in range(n_sites_1d):
        chosen_coin_op = coin0_op if current_ca_pattern[site_idx] == 0 else coin1_op
        idx_s_c0 = get_1d_index(0, site_idx, n_sites_1d)
        idx_s_c1 = get_1d_index(1, site_idx, n_sites_1d)
        C_full[idx_s_c0, idx_s_c0] = chosen_coin_op[0,0]
        C_full[idx_s_c1, idx_s_c0] = chosen_coin_op[1,0]
        C_full[idx_s_c0, idx_s_c1] = chosen_coin_op[0,1]
        C_full[idx_s_c1, idx_s_c1] = chosen_coin_op[1,1]
    S_full = np.zeros((state_dim_1d, state_dim_1d), dtype=np.complex128)
    for k_in in range(state_dim_1d):
        coin_val, site_pos = get_1d_coin_pos_from_index(k_in, n_sites_1d)
        new_pos = site_pos
        if coin_val == 0:
            new_pos = (site_pos - 1 + n_sites_1d) % n_sites_1d
        else:
            new_pos = (site_pos + 1) % n_sites_1d
        k_out = get_1d_index(coin_val, new_pos, n_sites_1d)
        S_full[k_out, k_in] = 1.0
    U_step = S_full @ C_full
    return U_step

# --- Observables (1D QW) ---
def calculate_p1_position_1d(state_vector, n_sites_1d):
    prob_dist_1d = np.zeros(n_sites_1d)
    for site_idx in range(n_sites_1d):
        idx_c0 = get_1d_index(0, site_idx, n_sites_1d)
        idx_c1 = get_1d_index(1, site_idx, n_sites_1d)
        prob_dist_1d[site_idx] = np.abs(state_vector[idx_c0])**2 + np.abs(state_vector[idx_c1])**2
    return prob_dist_1d

def calculate_coin_position_entanglement_1d(state_vector, n_sites_1d):
    coin_dim = 2
    state_dim_1d = coin_dim * n_sites_1d
    if len(state_vector) != state_dim_1d: return np.nan
    norm = np.linalg.norm(state_vector)
    if np.abs(norm - 1.0) > 1e-6:
        if norm < 1e-9: return 0.0
        state_vector = state_vector / norm
        if np.isnan(state_vector).any(): return np.nan
    try:
        rho_full_matrix = np.outer(state_vector, np.conjugate(state_vector))
        rho_coin = np.zeros((coin_dim, coin_dim), dtype=np.complex128)
        for site_idx in range(n_sites_1d):
            for c_row in range(coin_dim):
                 for c_col in range(coin_dim):
                      k_row = get_1d_index(c_row, site_idx, n_sites_1d)
                      k_col = get_1d_index(c_col, site_idx, n_sites_1d)
                      rho_coin[c_row, c_col] += rho_full_matrix[k_row, k_col]
        trace_rho_coin = np.trace(rho_coin)
        if np.abs(trace_rho_coin - 1.0) > 1e-6:
            if abs(trace_rho_coin) > 1e-9: rho_coin /= trace_rho_coin
        eigenvalues = np.linalg.eigvalsh(rho_coin)
        entropy = 0.0; epsilon = 1e-12
        for eig_val in eigenvalues:
            if eig_val > epsilon:
                try: entropy -= eig_val * np.log2(eig_val)
                except ValueError: pass
        return max(0.0, np.real(entropy))
    except Exception as e:
        # print(f"Error calculating 1D coin-position entanglement: {e}") # Reduce noise
        return np.nan

# --- Simulation Loop (1D QW with Two-Way Coupled Dynamic CA Coin) ---
def run_1d_qw_two_way_coupling(
    n_sites_1d, depth, initial_pos, initial_coin_val,
    initial_ca_center_one, ca_rule_number, coin0_op, coin1_op,
    p_threshold, feedback_type
):
    current_qw_state = prepare_initial_state_1d(n_sites_1d, initial_pos, initial_coin_val)
    current_ca_line = np.zeros(n_sites_1d, dtype=int)
    if initial_ca_center_one and n_sites_1d > 0:
        if n_sites_1d % 2 == 1: current_ca_line[n_sites_1d // 2] = 1
        else: current_ca_line[n_sites_1d // 2 -1] = 1

    prob_history_1d = [calculate_p1_position_1d(current_qw_state, n_sites_1d)]
    ent_history_1d = [calculate_coin_position_entanglement_1d(current_qw_state, n_sites_1d)]
    ca_history = [current_ca_line.copy()]

    # print_feedback_info = True # Flag to control feedback summary print (not used currently)

    start_time = time.time()
    for step in range(depth):
        U_step = build_1d_qw_step_operator(n_sites_1d, current_ca_line, coin0_op, coin1_op)
        current_qw_state = U_step @ current_qw_state
        norm_qw = np.linalg.norm(current_qw_state)
        if abs(norm_qw) > 1e-9 : current_qw_state /= norm_qw
        else: print(f"Warning: QW State norm zero at step {step+1}"); break

        qw_probs_at_sites = calculate_p1_position_1d(current_qw_state, n_sites_1d)
        ca_line_for_next_update = current_ca_line.copy()

        num_influenced_cells_this_step = 0
        for site_idx in range(n_sites_1d):
            if qw_probs_at_sites[site_idx] > p_threshold:
                num_influenced_cells_this_step +=1
                if feedback_type == "flip":
                    ca_line_for_next_update[site_idx] = 1 - ca_line_for_next_update[site_idx]
                elif feedback_type == "set_to_one":
                    ca_line_for_next_update[site_idx] = 1
                elif feedback_type == "set_to_zero":
                    ca_line_for_next_update[site_idx] = 0

        current_ca_line = update_ca_line(ca_line_for_next_update, ca_rule_number)

        prob_history_1d.append(qw_probs_at_sites)
        ent_history_1d.append(calculate_coin_position_entanglement_1d(current_qw_state, n_sites_1d))
        ca_history.append(current_ca_line.copy())

        if (step + 1) % (depth // 20 if depth >=100 else 10) == 0 :
            # print(f"  Step {step + 1}/{depth} completed. CA cells influenced in this step: {num_influenced_cells_this_step}") # Reduce noise
            pass

    end_time = time.time()
    print(f"Evolution complete. Time: {end_time - start_time:.2f} seconds.")

    return {
        "prob_history": np.array(prob_history_1d),
        "ent_history": np.array(ent_history_1d),
        "ca_history": np.array(ca_history),
        "params": {
            "n_sites": n_sites_1d, "depth": depth,
            "initial_pos": initial_pos, "initial_coin": initial_coin_val,
            "initial_ca_center_one": initial_ca_center_one,
            "ca_rule_number": ca_rule_number,
            "p_threshold": p_threshold,
            "feedback_type": feedback_type
        }
    }

# --- Plotting and Summary (1D QW with Dynamic CA and Two-Way Coupling) ---
def plot_1d_qw_results_two_way(results):
    params = results["params"]; prob_hist = results["prob_history"]
    ent_hist = results["ent_history"]; ca_hist = results["ca_history"]
    depth = params['depth']; ca_rule_number = params['ca_rule_number']
    p_thresh = params['p_threshold']; feedback = params['feedback_type']
    c0_name = params.get("coin0_name", "C0_def")
    c1_name = params.get("coin1_name", "C1_def")

    fig, axs = plt.subplots(4, 1, figsize=(12, 14), gridspec_kw={'height_ratios': [1, 3, 3, 2]})
    title_str = (f"1D QW Two-Way Coupled with Rule {ca_rule_number} CA\n"
                 f"(N={params['n_sites']}, D={depth}, Pth={p_thresh}, Fb='{feedback}', C0='{c0_name}', C1='{c1_name}')")
    fig.suptitle(title_str, fontsize=12)

    axs[0].imshow(ca_hist, cmap='binary', aspect='auto', interpolation='nearest')
    axs[0].set_title(f"Dynamic Rule {ca_rule_number} Pattern (Influenced by QW)")
    axs[0].set_xlabel("Site Index"); axs[0].set_yticks([])

    axs[1].plot(np.arange(params['n_sites']), prob_hist[-1, :], marker='o', linestyle='-', color='blue')
    axs[1].set_title("Final QW Probability Distribution P(x)")
    axs[1].set_xlabel("Site Index (x)"); axs[1].set_ylabel("Probability P(x)")
    axs[1].grid(True); axs[1].set_ylim(bottom=-0.01)

    prob_spacetime = prob_hist.T
    pos_probs = prob_spacetime[prob_spacetime > 1e-9]
    min_val = pos_probs.min() if len(pos_probs) > 0 else 1e-9
    max_val = prob_spacetime.max()
    norm_val = colors.LogNorm(vmin=min_val, vmax=max(max_val, min_val + 1e-6))
    im = axs[2].imshow(prob_spacetime, aspect='auto', origin='lower', cmap='magma', norm=norm_val,
                       extent=[0, depth, 0, params['n_sites']-1])
    plt.colorbar(im, ax=axs[2], label="P(x,t) (log scale)")
    axs[2].set_title("QW Probability Spacetime P(x,t)")
    axs[2].set_xlabel("QW Time Step (t)"); axs[2].set_ylabel("Site Index (x)")

    time_steps_qw = np.arange(depth + 1)
    axs[3].plot(time_steps_qw, ent_hist, marker='.', linestyle='-', color='green')
    axs[3].set_title("Coin-Position Entanglement Entropy S(t)")
    axs[3].set_xlabel("QW Time Step (t)"); axs[3].set_ylabel("Entropy S")
    axs[3].grid(True); axs[3].set_ylim(bottom=-0.01, top=1.05)

    plt.tight_layout(rect=[0, 0, 1, 0.92]);
    plt.show()

def generate_1d_qw_text_output_two_way(results):
    params = results["params"]; prob_hist = results["prob_history"]; ent_hist = results["ent_history"]
    ca_rule_number = params['ca_rule_number']
    p_thresh = params['p_threshold']; feedback = params['feedback_type']
    c0_name = params.get("coin0_name", "C0_def")
    c1_name = params.get("coin1_name", "C1_def")

    output_lines = [f"\n--- 1D QW (Two-Way Coupled Rule {ca_rule_number}, Pth={p_thresh}, Fb='{feedback}', C0='{c0_name}', C1='{c1_name}') Summary ---"]
    output_lines.append(f"N_sites={params['n_sites']}, Depth={params['depth']}, StartPos={params['initial_pos']}, StartCoin={params['initial_coin']}")

    output_lines.append("\nTop 5 Final Probabilities (Site: Prob):")
    final_probs = prob_hist[-1, :]
    sorted_indices = np.argsort(final_probs)[::-1]
    for i in range(min(5, params['n_sites'])):
        idx = sorted_indices[i]
        output_lines.append(f"  Site {idx}: {final_probs[idx]:.5f}")

    output_lines.append("\nCoin-Position Entanglement vs. Time:")
    num_steps_total = len(ent_hist)
    indices_to_show = [0]
    if num_steps_total > 2: indices_to_show.append(num_steps_total // 2)
    if num_steps_total > 1: indices_to_show.append(num_steps_total - 1)
    indices_to_show = sorted(list(set(indices_to_show)))
    for idx in indices_to_show:
        s_val_str = f"{ent_hist[idx]:.6f}" if np.isfinite(ent_hist[idx]) else "nan"
        output_lines.append(f"  Step {idx}: S = {s_val_str}")
    output_lines.append("--- End of Summary ---")
    return "\n".join(output_lines)


# --- Main Execution (1D QW - Specific Long Run) ---
if __name__ == "__main__":

    all_results_text = []

    # --- Experiment: Long Evolution with Rule 110, C0=H, C1=I ---
    long_run_depth = 5000 # Increased depth
    long_run_p_thresh = 0.05
    long_run_feedback = "flip"
    long_run_rule = 110
    long_run_c0 = H_1Q    # COIN_OP_CA_ZERO
    long_run_c1 = I_1Q    # COIN_OP_CA_ONE set to Identity

    def get_op_name(op_matrix):
        if np.allclose(op_matrix, H_1Q): return "H"
        if np.allclose(op_matrix, X_1Q): return "X"
        if np.allclose(op_matrix, I_1Q): return "I"
        if np.allclose(op_matrix, X_1Q @ H_1Q): return "XH"
        return "Custom"

    c0_name_long = get_op_name(long_run_c0)
    c1_name_long = get_op_name(long_run_c1)

    experiment_title = f"Rule {long_run_rule}, D={long_run_depth}, Pth={long_run_p_thresh}, Fb='{long_run_feedback}', C0='{c0_name_long}', C1='{c1_name_long}'"
    print(f"\n===== STARTING LONG RUN EXPERIMENT: {experiment_title} =====")
    print(f"Initialized 1D walker at site {INITIAL_POS_1D} in coin state |{INITIAL_COIN_1D}>") # Moved init print here

    results_long_run = run_1d_qw_two_way_coupling(
        n_sites_1d=N_SITES_1D,
        depth=long_run_depth,
        initial_pos=INITIAL_POS_1D,
        initial_coin_val=INITIAL_COIN_1D,
        initial_ca_center_one=INITIAL_CA_CENTER_ONE,
        ca_rule_number=long_run_rule,
        coin0_op=long_run_c0,
        coin1_op=long_run_c1,
        p_threshold=long_run_p_thresh,
        feedback_type=long_run_feedback
    )
    results_long_run["params"]["coin0_name"] = c0_name_long
    results_long_run["params"]["coin1_name"] = c1_name_long

    text_summary_long = generate_1d_qw_text_output_two_way(results_long_run)
    all_results_text.append(f"--- Long Run Experiment ({experiment_title}) Results ---\n{text_summary_long}")
    print(text_summary_long)
    plot_1d_qw_results_two_way(results_long_run)
    print(f"===== Finished Long Run Experiment =====")


    print("\n\n<<<<< ALL SIMULATION SUMMARIES >>>>>")
    for summary in all_results_text:
        print(summary)
    print("<<<<< END OF ALL SIMULATION SUMMARIES >>>>>")
```
